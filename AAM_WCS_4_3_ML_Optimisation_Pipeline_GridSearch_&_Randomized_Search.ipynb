```python
# Importation de la Biblioth√®que pour la lecture/traitement de DF
import pandas as pd
import numpy as np

# Importe ce dataset relatif au Titanic dans un DataFrame.
link = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df_titanic = pd.read_csv(link)
df_titanic.head()
```

```python
# Afin de s'assurer un nombre conc√©quent de variables explicatives possibles
# Ajouter une colonne sexe en num√©rique par la m√©thode ".factorize()[0]"
df_titanic['sex_numeric'] = df_titanic['Sex'].factorize()[0]
df_titanic
```

1) Tu vas effectuer une classification supervis√©e sur la colonne "survived", en utilisant le classifieur **decision tree**. Tu dois faire apparaitre clairement l'**accuracy** score du jeu de test.

```python
# importons les biblioth√®ques
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier


X = df_titanic.select_dtypes(include=np.number) # le ".select_dtypes(include=np.number)" s'assure de pratiquer le ML sur les variables num√©riques du DF
y = df_titanic["Survived"]

# Diviser les donn√©es
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Entra√Ænement du mod√®le "arbre de d√©cision"
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Pr√©dire sur l'ensemble de test
y_pred = clf.predict(X_test)

# Calculer et afficher le score de la pr√©cision du jeux test
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# RMQ: Bingo!!! Score de pr√©cision trop parfait ?!
```

```python
# Afin de corriger le tire , assignons mieux le X & y en supprimant la variable qui est suspect de bias√© notre ML
X = df_titanic.select_dtypes(include='number').drop(columns=['Survived'])
y = df_titanic['Survived']

# re-entra√Ænons le model de classification "clt" de l'"arbre de d√©cision"
X_train, X_test, y_train, y_test = train_test_split(X,y)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

print("Accuracy:", accuracy)

# RMQ: Bof!, pas tr√®s grand changement sur l'Accuracy !
```

```python
#Afficher les param√®tres de bases du mod√®le "arbre de d√©cision"
clf.get_params()
```

```python
# Voir la matrice de confusion pour les donn√©es de test et les pr√©dictions

# import des 3 m√©triques du ML
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

confusion_matrix(y_test, y_pred)
```

```python
# Voir le "classification_report" pour les donn√©es de test et les pr√©dictions

classification_report(y_test, y_pred)
```

```python
# Obtenir le score du train et du test
tree_train_score = clf.score(X_train, y_train)
tree_test_score = clf.score(X_test, y_test)
score_tree_df = pd.DataFrame({'train score' : [tree_train_score],
                              'test score' : [tree_test_score]})
score_tree_df

# RMQ: les scores semblent correctes!
```

```python
# Imprimer chaque colonne de variables explicatives avec son importance

importance_df = pd.DataFrame(clf.feature_importances_,
                             X.columns).sort_values(by=0, ascending=False).T
importance_df

# RMQ : le sexe a plus tendance √† expliquer la survie/ ou non sur le Titanic !
```

2) Avec ce m√™me classifieur, tu vas lancer une **GridSearch** avec les valeurs :
* tous les entiers de 1 √† 50 du param√®tre max_depth tous les entiers entre 1 et
* 15 du param√®tre min_samples_leaf
* (2, 5, 7, 10, 15, 30) du param√®tre min_samples_split

```python
# import de la librairie GridSearch
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': range(1,51),
              'min_samples_leaf': range(1,16),
              'min_samples_split': [2, 5, 7, 10, 15, 30]
             }
```

/ ! \ Le traitement ci-dessous üëáüèΩ devrait durer quelques minutes

```python
# Effectuer le Cross-Validation avec GridSearch pour la Classification
grid_search = GridSearchCV(clf, param_grid, scoring='accuracy')
grid_search.fit(X,y)
```

```python
# Afficher tout les param√®tres de bases du mod√®le de cross-validation avec la m√©thode "GridSearch"
grid_search.get_params()
```

```python
# Afficher le r√©sultat de la "GridSearch" dans un DF
import pandas as pd
grid_cv_df = pd.DataFrame(grid_search.cv_results_)
grid_cv_df
```

**/ ! \ Attention**, le traitement ci-dessous üëáüèΩ devrait durer quelques minutes, car GridSearch doit tester 4500 combinaisons de param√®tres ! En effet, les combinaisons de param√®tres sont 50 * 15 * 6 = 4500

3) Avec cette recherche via GridSearch, quelles sont le meilleures valeurs des **hyperparam√®tres** si on cherche √† maximiser l'**accuracy score** ?

```python
# Entra√Ænement du mod√®le GridSearch
grid_search.fit(X_train, y_train)

# Afficher les "best" param√®tres pris en compte pour un "parfait" l'Accuracy
print("Les meilleures parametr√®s sont :", grid_search.best_params_)
print("Le meilleure Score de pr√©cision du GridSearch (Accuracy) est:", grid_search.best_score_)
```

```python
# Best Estimateur du GridSearch est...
grid_search.best_estimator_

# ...c'est pour un profondeur √©gale √† 1 ("max_depth=1" ): l'arbre de d√©cision
```

4) Avec ce m√™me classifieur, tu vas lancer un **RandomSearch** sur les m√™mes valeurs des param√®tres en limitant √† 200 it√©rations.

Cette fois, le traitement devrait durer quelques secondes, car nous avons limit√© √† 200 it√©rations.

```python
# import de la librairie de RandomizedSearch
from sklearn.model_selection import RandomizedSearchCV
# import la/ de(s) librairie des(s) m√©trique(s) associ√©(s) au RandomizedSearch
from scipy.stats import randint as sp_randint
# import de la librairie de Classification KNN
from sklearn.neighbors import KNeighborsClassifier

param_dist = {'max_depth': list(range(1,51)),
              'min_samples_leaf': list(range(1,16)),
              'min_samples_split': [2, 5, 7, 10, 15, 30]
             }
```

```python
# RandomizedSearch avec random forest Classifier:

#import des librairies
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
RDSCV = RandomizedSearchCV(model, param_dist ,scoring="accuracy")
RDSCV.fit(X_train,y_train)
RDSCV.best_params_
best_model = RDSCV.best_estimator_
```

```python
# les param√®tres du best model de RandomizedSearch
best_model
```

```python
#  Les m√©triques du best mod√®le: accuracy
y_pred_test = best_model.predict(X_test)
print(accuracy_score(y_test, y_pred_test))
```

```python
# Les m√©triques du best mod√®le : matrice de confusion
confusion_matrix(y_test, y_pred_test)
```

5) Avec cette recherche via RandomSearch, quelles sont le meilleures valeurs des hyperparam√®tres si on cherche √† maximiser l'accuracy score ?
Les param√®tres optimaux peuvent √™tre diff√©rents entre les deux m√©thodes. En effet :

* en randomsearch, tous les param√®tres ne sont pas test√©s

* si les scores sont tr√®s proches entre plusieurs combinaisons de param√®tres. Ici, est-ce bien le cas ? Les param√®tres sont-ils identiques entre grid et random ? Sinon, les scores sont-ils bien tr√®s proches ?


```python
# Effectuer le Cross-Validation avec RandomizedSearch pour la Classification
random_search = RandomizedSearchCV(clf, param_distributions=param_dist, n_iter=200, scoring='accuracy')

# test plus "classique" c_a_d "train" avec le mod√®le RandomizedSearch
random_search.fit(X_train, y_train)

# Afficher les "best" param√®tres pris en compte pour un "parfait" l'Accuracy
print("Les meilleures parametr√®s sont :", random_search.best_params_)
print("Le meilleure Score de pr√©cision du RandomizedSearch (Accuracy) est:", random_search.best_score_)
```

```python
# Afficher les resultats dans un DF
rando_cv_df = pd.DataFrame(random_search.cv_results_)
rando_cv_df
```
