```python
POKEMON=["Mewtwo", "Lugia", "Rayquaza", "Giratina Origin Forme", "Dialga", "Palkia"]
for pokemon in POKEMON:              #  Je fais un try except parce que certains pokemon ne sont pas reconnu avec un simple '=='. Du coup je le fais aller dans le except avec le str.contains
  try:
    ok=model.kneighbors(legendary.loc[legendary['Name'] == pokemon, X.columns])
  except:
    ok=model.kneighbors(legendary.loc[legendary['Name'].str.contains(pokemon), X.columns])
  ok2=ok[1][0][0]
  ok2=df['Name'].iloc[ok2]
  print(f"Le pokemon le plus proche de {pokemon} est {ok2}")
```
