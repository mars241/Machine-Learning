```python
import numpy as np
import pandas as pd
#import seaborn as sns
from sklearn.neighbors import NearestNeighbors

link = "https://raw.githubusercontent.com/murpi/wilddata/master/pokemon.csv"
df_pokemons = pd.read_csv(link, sep=',',na_values=['NA','\\N'], index_col=0)
df_pokemons.head()
```

```python
df_pokemons.dropna(axis=1, inplace=True)
```

```python
df_pokemons.info()
```

```python
df_pokemons[df_pokemons['Name'] == 'Giratina']
```

```python
df_pokemons[df_pokemons['Name'].isin(['Mewtwo','Lugia', 'Rayquaza', 'Dialga', 'Palkia'])]
```

```python
pokemons_non_legend = df_pokemons[df_pokemons.Legendary == False]
pokemons_legend = df_pokemons[df_pokemons.Legendary == True]
```

```python
# Entraine l'algorithme NN en utilisant l'ensemble des colonnes numériques en variables explicatives X.

X = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Generation']

distanceKNN = NearestNeighbors(n_neighbors=1).fit(pokemons_non_legend[X])
model = pokemons_legend[pokemons_legend['Name'] =='Mewtwo'][X]
distanceKNN.kneighbors(model)
```

```python
neighbors = distanceKNN.kneighbors(model)
pokemons_non_legend.iloc[neighbors[1][0][0]]
```

```python
# une boucle pour trouver tout les remplaçants (Pokémon non légendaires)

pokemon_champion = ['Mewtwo','Lugia', 'Rayquaza', 'Dialga', 'Palkia']
for i in pokemon_champion:
  model = pokemons_legend[pokemons_legend['Name'] == i][X]
  distanceKNN.kneighbors(model)
  neighbors = distanceKNN.kneighbors(model)

  print(f"Le Pokémon non légendaires pour la prochaine compétition est : {pokemons_non_legend.iloc[neighbors[1][0][0]].Name}")
```
